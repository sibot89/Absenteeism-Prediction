The ID column is not usefull, so we will drop it.

```python
df = df.drop(['ID'], axis=1)
df.head()
```

Now, we want to check the column "Reason for Absence".
This is a categorical column, so we should create dummy variables for it; but first we must see the number of categories.

```python
df['Reason for Absence'].min(), df['Reason for Absence'].max()
```

```python
type(df['Reason for Absence'])
```

```python
type(df['Reason for Absence'][0])
```

So the category is consist of integers between 0 and 28. The next question is that are they all used? or some of the numbers is missing? Let's check it!

```python
len(pd.unique(df['Reason for Absence']))
```

The category is in range(0,28), that is consist of the 29 individual categories, but we have 28 unique number. So one category is missing. Wich one? Let's find!

```python
sorted(pd.unique(df['Reason for Absence']))
```

```python
def find_missing(categories):
    index = -1
    for i in range(len(categories)-1):
        if categories[i+1] - categories[i] > 1:
            index = i+1
    # print(f"index is: {index}")
    return index
```

```python
find_missing(sorted(pd.unique(df['Reason for Absence'])))
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], dtype=int)
reason_columns
```

```python
reason_columns['Ceck'] = reason_columns.sum(axis=1)
reason_columns
```

Here, we add a check column to our dummy variables to check them. This column apply sum function on every row,
we must have one and only one number 1 in a row and the other numbers should be 0.
But it's hard to found that is this condition satisfied or not just by eyeballing.
Let's check it more precisely!

```python
reason_columns['Ceck'].sum(axis=0)
```

```python
reason_columns['Ceck'].unique()
```

So we have one and only one number 1 in a row and the other numbers are 0.
Now, we don't need check column any more. Also we shuld drop one column to avoid Multicollinearity.
The next line satisfied these.

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], dtype=int, drop_first=True)
reason_columns
```

```python
The dummy variables are created, but this count of variables are too larg for our data set. 
To reduce the feature space we group by reasons in 4 category.
```

```python
reason_type1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type3 = reason_columns.loc[:,18:22].max(axis=1)
reason_type4 = reason_columns.loc[:,23:].max(axis=1)
```

Now we drop column 'Reason for Absence' and add this new columns.

```python
df = df.drop(['Reason for Absence'], axis=1)
df.head()
```

```python
df = pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis=1)
df.head()
```

Then we choose better name for our new features and move them into first of the data frame.

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average(in Minutes)', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason1', 'Reason2', 'Reason3', 'Reason4']
df.columns = column_names
df.head()
```

```python
reordered_columns = ['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average(in Minutes)', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[reordered_columns]
df.head()
```

Creating a Checkpoint

```python
df_reason_modified = df.copy()
df_reason_modified.head()
```

Now we want to consider 'Date' column.
Here we want to care about the month and the day of week for absenteeism.
But first of all, let's check the type.

```python
type(df_reason_modified['Date'])
```

```python
type(df_reason_modified['Date'][0])
```

The type of 'Date' column is string. Here we convert it to Date

```python
df_reason_modified['Date'] = pd.to_datetime(df_reason_modified['Date'])
type(df_reason_modified['Date'][0])
```

```python
def date_to_month(date_value):
    return date_value.month
```

```python
df_reason_modified['Month'] = df_reason_modified['Date'].apply(date_to_month)
df_reason_modified.head()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_modified['Day of the Week'] = df_reason_modified['Date'].apply(date_to_weekday)
df_reason_modified.head()
```

```python
df_reason_modified = df_reason_modified.drop(['Date'] , axis=1)
```

```python
df_reason_modified.columns.values
```

```python
reordered_names = ['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Month', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average(in Minutes)', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_modified = df_reason_modified[reordered_names]
df_reason_modified.head()
```

Creating the Checkpoint

```python
df_reason_date_modified = df_reason_modified.copy()
df_reason_date_modified.head()
```

The columns 'Transportation Expense', 'Distance to Work', 'Daily Work Load Average(in Minutes)' and Body Mass Index	will use with
any change.
Now we want to analys the column 'Education'

```python
df_reason_date_modified['Education'].unique()
```

```python
df_reason_date_modified['Education'].value_counts()
```

As we can see, the number of members in each category is very diffrent => It's better to devide it to 2 categories
consist of 1 and (2,3,4)

```python
df_reason_date_modified['Education'] = df_reason_date_modified['Education'].map({1:0, 2:1, 3:1, 4:1})
df_reason_date_modified['Education'].unique()
```

```python
df_reason_date_modified['Education'].value_counts()
```

Final checkpoint and Exporting our Preprocessed Dataset

```python
df_preprocessed = df_reason_date_modified.copy()
df_preprocessed.head()
```
